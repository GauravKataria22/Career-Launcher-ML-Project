```python
#Problem statment 2.9
#To do-
# 1. Plot the 21 day and 34 day moving average with the average price and decide a call!
```

```python
import numpy as np
short_window = 21
long_window = 34
signals = pd.DataFrame(index=bajaj.index)
signals['signal'] = 0.0
signals['short_mavg'] = bajaj['Close Price'].rolling(window=short_window, min_periods=1,center=False).mean()
signals['long_mavg'] = bajaj['Close Price'].rolling(window=long_window,min_periods=1, center=False).mean()
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1.0,0.0)
signals['positions'] = signals['signal'].diff()
print(signals)
```

```python
fig = plt.figure(figsize=(17,5))
ax1 = fig.add_subplot(111, ylabel='Price')
bajaj['Close Price'].plot(ax=ax1, color='green', lw=2.)
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short_mavg[signals.positions == 1.0], '^' , markersize=20,color='b')
ax1.plot(signals.loc[signals.positions == -1.0].index, signals.short_mavg[signals.positions == -1.0], 'v' , markersize=20,color='r')
plt.show()
```

```python
#Problem Statement 2.10
#To do-
# 1. Plot the bollinger bands for this stock
```

```python
symbol = 'Nifty'
df = pd.read_csv(r'C:\Users\gaura\Desktop\Nifty50.csv'.format(symbol), index_col='Date',
                 parse_dates=True, usecols=['Date', 'Close'],
                 na_values='nan')
df = df.rename(columns={'Close': symbol})
df.dropna(inplace=True)
sma = df.rolling(window=14).mean()
rstd = df.rolling(window=14).std()
upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})
df = df.join(upper_band).join(lower_band)
ax = df.plot(title='{} Price'.format(symbol))
ax.fill_between(df.index, lower_band['lower'], upper_band['upper'], color='#adffca', alpha='0.4')
ax.set_xlabel('Date')
ax.set_ylabel('')
ax.grid()
plt.show()
```
