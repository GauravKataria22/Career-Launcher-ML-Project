```python
#2. 
def pl_perc(N):
    diff=df['Close Price'][df.index[-N-1]] - df['Close Price'][df.index[-N]]
    return((diff/df['Close Price'][df.index[-N]])*100)
```

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change().fillna(0)
df
```

```python
#Problem Statement 1.7
#To do:
# 1. Add column Trend
```

```python
def label(row):
    if ((row['Day_Perc_Change'] >= -0.5) & (row['Day_Perc_Change'] <= 0.5)):
        df['Trend'] = 'Slight or No change'
    if ((row['Day_Perc_Change'] >= 0.5) & (row['Day_Perc_Change'] <= 1)):
        df['Trend'] = 'Slight positive'
    if ((row['Day_Perc_Change'] <= -0.5) & (row['Day_Perc_Change'] >= -1)):
        df['Trend'] = 'Slight negative'
    if ((row['Day_Perc_Change'] >= 1) & (row['Day_Perc_Change'] <= 3)):
        df['Trend'] = 'Positive' 
    if ((row['Day_Perc_Change'] <= -1) & (row['Day_Perc_Change'] >= -3)):
        df['Trend'] = 'Negative'
    if ((row['Day_Perc_Change'] >= 3) & (row['Day_Perc_Change'] <= 7)):
        df['Trend'] = 'Among top gainers'
    if ((row['Day_Perc_Change'] <= -3) & (row['Day_Perc_Change'] >= -7)):
        df['Trend'] = 'Among top losers'
    if (row['Day_Perc_Change'] > 7):
        df['Trend'] = 'Bull run' 
    if (row['Day_Perc_Change'] < -7):
        df['Trend'] = 'Bear drop' 
```

```python
df.apply (lambda row: label(row), axis=1)
```
